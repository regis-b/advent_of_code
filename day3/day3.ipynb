```python
import pandas as pd
import numpy as np

file_path = 'input.txt'
df = pd.read_csv(file_path, header=None, sep=' ')
df.columns = ['content']

df
```

```python
def first_half(s):
    return s[:len(s)//2]
    
def second_half(s):
    return s[len(s)//2:]
df['compartment1'] = df['content'].map(first_half)
df['compartment2'] = df['content'].map(second_half)
df
```

```python
def find_common_character(s1, s2):
    l1 = list(s1)
    l2 = list(s2)
    return list(set(l1).intersection(l2))[0]

df['in_both'] = df.apply(lambda x: find_common_character(x.compartment1, x.compartment2), axis=1)
pd.set_option('display.max_rows', 40)
df
```

```python
priority_dict = {
    'a': 1, 'b': 2, 'c': 3, 'd': 4, 'e': 5, 'f': 6, 'g': 7, 'h': 8, 'i': 9, 'j': 10,
    'k': 11, 'l': 12, 'm': 13, 'n': 14, 'o': 15, 'p': 16, 'q': 17, 'r': 18, 's': 19,
    't': 20, 'u': 21, 'v': 22, 'w': 23, 'x': 24, 'y': 25, 'z': 26, 'A': 27, 'B': 28,
    'C': 29, 'D': 30, 'E': 31, 'F': 32, 'G': 33, 'H': 34, 'I': 35, 'J': 36, 'K': 37,
    'L': 38, 'M': 39, 'N': 40, 'O': 41, 'P': 42, 'Q': 43, 'R': 44, 'S': 45, 'T': 46,
    'U': 47, 'V': 48, 'W': 49, 'X': 50, 'Y': 51, 'Z': 52,
    }
df['priority'] = df['in_both']
df.replace({"priority": priority_dict}, inplace=True)

df
```

```python
df2 = pd.read_csv(file_path, header=None, sep=' ')
df2.columns = ['content']
df2 = df2.reset_index()
df2
```

```python
df2['elf_number'] = df2['index'].mod(3)
df2['group_number'] = np.floor(df2['index'].div(3)).astype('int')
df2
```

```python
df3 = df2.pivot(index='group_number', columns='elf_number', values='content')
df3.columns = ['content1', 'content2', 'content3']
df3
```

```python
def find_common_character2(s1, s2, s3):
    l1 = list(s1)
    l2 = list(s2)
    l3 = list(s3)
    return list(set(l1).intersection(l2).intersection(l3))[0]

df3['common'] = df3.apply(lambda x: find_common_character2(x.content1, x.content2, x.content3), axis=1)

df3['priority'] = df3['common']
df3.replace({"priority": priority_dict}, inplace=True)
df3
```
