```python
import pandas as pd
import numpy as np

file_path = 'input.txt'
df = pd.read_csv(file_path, header=None, sep=',')
df.columns = ['elf1', 'elf2']

# Find range min and max of each elf
df[['elf1_min','elf1_max']] = df['elf1'].str.split('-',expand=True)
df[['elf2_min','elf2_max']] = df['elf2'].str.split('-',expand=True)
df['elf1_min'] = df['elf1_min'].astype('int')
df['elf1_max'] = df['elf1_max'].astype('int')
df['elf2_min'] = df['elf2_min'].astype('int')
df['elf2_max'] = df['elf2_max'].astype('int')
df

# Find rows with range included in the other
cond = (
    (
        (df['elf1_min'] >= df['elf2_min'])
        & (df['elf1_max'] <= df['elf2_max'])
    )
    |(
        (df['elf2_min'] >= df['elf1_min'])
        & (df['elf2_max'] <= df['elf1_max'])
    )
)

df[cond]
```

```python
# search for overlaping ranges
df['min_of_max'] = df[['elf1_max', 'elf2_max']].min(axis=1)
df['max_of_min'] = df[['elf1_min', 'elf2_min']].max(axis=1)
df[df['max_of_min'] <= df['min_of_max']]
```
