```python
import numpy as np
# input_file = 'input_example.txt'
input_file = 'input_full.txt'

with open(input_file) as file:
    lines = file.readlines()
jets = list(lines[0])
# print(lines[0])
```

```python
hline =  [complex(0,0), complex(1,0), complex(2,0), complex(3,0)]
cross =  [complex(1,0), complex(0,1), complex(1,1), complex(2,1), complex(1,2)]
L =      [complex(0,0), complex(1,0), complex(2,0), complex(2,1), complex(2,2)]
vline =  [complex(0,0), complex(0,1), complex(0,2), complex(0,3)]
square = [complex(0,0), complex(1,0), complex(0,1), complex(1,1)]
shapes = [hline, cross, L, vline, square]
```

```python
def print_shape(shape):
    for y in range(3,-1,-1):
        shape_str = ''
        for x in range(0,4):
            if complex(x,y) in shape:
                shape_str += '@'
            else:
                shape_str += '.'
        print(shape_str)

print_shape(shapes[1])
```

```python
import pandas as pd
# We log the height and the shape nb. The idea is to find a pattern to extrapolate.
df = pd.DataFrame()
def log_height(shape_nbr, height):
    global df
    new_row = pd.DataFrame({'shape_nbr': shape_nbr, 'height': height}, index=[0])
    df =  pd.concat([df, new_row]).reset_index(drop=True)

# Initialization: Add a first row of rocks at the bottom, ...
rocks = [complex(n,0) for n in range(1,8)]
shape_nbr = -1
nb_landed = 0
need_new_shape = True
for time in range(0,10000000000000000):
    # If a new shape is needed, we initialize it
    if need_new_shape:
        shape_nbr = (shape_nbr+1)%5
        shape = shapes[shape_nbr]
        rocks_top = np.array(rocks).imag.max()
        shape_xy = complex(3,rocks_top+4) # Bottom left coord of the shape
        need_new_shape = False
    # Apply jet: simulate where the shape would be after jet, and check if it s possible
    jet = jets[time%len(jets)]
    move = complex(1,0) if jet == '>' else complex(-1,0)
    shape_xy_attempt = shape_xy + move 
    can_move = True
    for xy in shape:
        xy_attempt = shape_xy_attempt + xy
        if (xy_attempt.real > 7) or (xy_attempt.real < 1) or (xy_attempt in rocks):
            can_move = False
            break
    if can_move:
        shape_xy = shape_xy_attempt
    
    # Test whether the shape can move down or must land
    shape_xy_attempt = shape_xy + complex(0,-1)
    must_land = False
    for xy in shape:
        if(shape_xy_attempt + xy) in rocks:
            must_land = True
            break
    if must_land:
        for xy in shape:
            rocks += [shape_xy + xy]
        need_new_shape = True
        nb_landed += 1
        log_height(nb_landed, np.array(rocks).imag.max())
        if nb_landed == 3000:
            break
    else:
        shape_xy = shape_xy_attempt
print(f'Part 1: {np.array(rocks).imag.max()}')
```
