```python
import pandas as pd
import numpy as np

file_path = 'input.txt'
df = pd.read_csv(file_path, header=None, sep=' ')
df.columns = ['opponent', 'me']

letters_to_shapes = {
    'A': 'rock',
    'B': 'paper',
    'C': 'scissors',
    'X': 'rock',
    'Y': 'paper',
    'Z': 'scissors',
}
shape_values = {
    'rock': 1,
    'paper': 2,
    'scissors': 3,
}

# Replace letters by shape name
df['opponent_shape'] = df['opponent']
df['me_shape'] = df['me']
df.replace({'opponent_shape': letters_to_shapes, 'me_shape': letters_to_shapes}, inplace=True)
# Replace shapes by points
df['me_shape_gain'] = df['me_shape']
df.replace({'me_shape_gain': shape_values}, inplace=True)
# Calculate battle gain
df['me_battle_gain'] = np.where(df['opponent_shape'] == df['me_shape'], 3, None)
df['me_battle_gain'] = np.where((df['opponent_shape'] == 'rock')     & (df['me_shape'] == 'paper'),    6, df['me_battle_gain'])
df['me_battle_gain'] = np.where((df['opponent_shape'] == 'rock')     & (df['me_shape'] == 'scissors'), 0, df['me_battle_gain'])
df['me_battle_gain'] = np.where((df['opponent_shape'] == 'paper')    & (df['me_shape'] == 'rock'),     0, df['me_battle_gain'])
df['me_battle_gain'] = np.where((df['opponent_shape'] == 'paper')    & (df['me_shape'] == 'scissors'), 6, df['me_battle_gain'])
df['me_battle_gain'] = np.where((df['opponent_shape'] == 'scissors') & (df['me_shape'] == 'rock'),     6, df['me_battle_gain'])
df['me_battle_gain'] = np.where((df['opponent_shape'] == 'scissors') & (df['me_shape'] == 'paper'),    0, df['me_battle_gain'])
# Calculate total gain
df['me_total_gain'] = df['me_shape_gain'] + df['me_battle_gain']

df
```

```python
letters_to_objective = {
    'X': 'lose',
    'Y': 'draw',
    'Z': 'win',
}
df['me_objective'] = df['me']
df.replace({'me_objective': letters_to_objective}, inplace=True)

# Mapping between opponent_shape and ovjective
df['me_shape2'] = np.where(df['me_objective'] == 'draw', df['opponent_shape'], None)
df['me_shape2'] = np.where((df['opponent_shape'] == 'rock')     & (df['me_objective'] == 'lose'), 'scissors', df['me_shape2'])
df['me_shape2'] = np.where((df['opponent_shape'] == 'rock')     & (df['me_objective'] == 'win'),  'paper'   , df['me_shape2'])
df['me_shape2'] = np.where((df['opponent_shape'] == 'paper')    & (df['me_objective'] == 'lose'), 'rock'    , df['me_shape2'])
df['me_shape2'] = np.where((df['opponent_shape'] == 'paper')    & (df['me_objective'] == 'win'),  'scissors', df['me_shape2'])
df['me_shape2'] = np.where((df['opponent_shape'] == 'scissors') & (df['me_objective'] == 'lose'), 'paper'   , df['me_shape2'])
df['me_shape2'] = np.where((df['opponent_shape'] == 'scissors') & (df['me_objective'] == 'win'),  'rock'    , df['me_shape2'])

# Calculate shape gain (reuse of above code)
df['me_shape_gain2'] = df['me_shape2']
df.replace({'me_shape_gain2': shape_values}, inplace=True)
# Calculate battle gain (reuse of above code)
df['me_battle_gain2'] = np.where(df['opponent_shape'] == df['me_shape2'], 3, None)
df['me_battle_gain2'] = np.where((df['opponent_shape'] == 'rock')     & (df['me_shape2'] == 'paper'),    6, df['me_battle_gain2'])
df['me_battle_gain2'] = np.where((df['opponent_shape'] == 'rock')     & (df['me_shape2'] == 'scissors'), 0, df['me_battle_gain2'])
df['me_battle_gain2'] = np.where((df['opponent_shape'] == 'paper')    & (df['me_shape2'] == 'rock'),     0, df['me_battle_gain2'])
df['me_battle_gain2'] = np.where((df['opponent_shape'] == 'paper')    & (df['me_shape2'] == 'scissors'), 6, df['me_battle_gain2'])
df['me_battle_gain2'] = np.where((df['opponent_shape'] == 'scissors') & (df['me_shape2'] == 'rock'),     6, df['me_battle_gain2'])
df['me_battle_gain2'] = np.where((df['opponent_shape'] == 'scissors') & (df['me_shape2'] == 'paper'),    0, df['me_battle_gain2'])
# Calculate total gain
df['me_total_gain2'] = df['me_shape_gain2'] + df['me_battle_gain2']

df
```
